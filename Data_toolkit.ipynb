THEORY QUESTIONS

.What is NumPy, and why is it widely used in Python?

ANS- NumPy is a powerful Python library for numerical computations. It provides support for arrays (specifically, multi-dimensional arrays), matrices, and many mathematical functions to operate on these data structures. It's widely used due to its speed (because it's implemented in C), its array broadcasting capabilities, and its extensive collection of mathematical functions, which makes it an essential tool for numerical, scientific, and machine learning tasks.

.How does broadcasting work in NumPy?

ANS-Broadcasting allows NumPy to perform element-wise operations on arrays of different shapes, so long as they are compatible. For instance, you can add a scalar value to a matrix, and NumPy will "broadcast" the scalar to each element of the matrix. The rule is that dimensions must either be the same or one of them must be 1.

.What is a Pandas DataFrame?

ANS-A DataFrame is a 2-dimensional labeled data structure in Pandas. It can store data of different types (e.g., integers, floats, strings) in columns. It's similar to a table in a database, an Excel spreadsheet, or a dictionary of Series. The key features of a DataFrame are its ability to handle missing data, perform groupby operations, and provide powerful indexing.

.Explain the use of the groupby() method in Pandas?

ANS- The groupby() method in Pandas is used to split the data into groups based on some criteria (like column values), perform a function (e.g., sum, mean), and then combine the results back into a DataFrame. It's commonly used for aggregation tasks such as finding the sum of a column for each group or the average of values within a group.

.Why is Seaborn preferred for statistical visualizations?

ANS-Seaborn is preferred for statistical visualizations because it simplifies complex visualization tasks by providing high-level interfaces to create beautiful and informative charts, like boxplots, violin plots, and heatmaps. It also integrates well with Pandas DataFrames, handles categorical data better, and includes automatic calculation of summary statistics.

.What are the differences between NumPy arrays and Python lists?

ANS-NumPy arrays are fixed-size, homogeneous data structures that support vectorized operations (i.e., applying an operation to every element efficiently). In contrast, Python lists are flexible, heterogeneous data structures that can store items of any type. NumPy arrays are much faster for numerical computations due to their compact memory layout and optimized operations.

.What is a heatmap, and when should it be used?

ANS-A heatmap is a graphical representation of data where values are depicted by color. It is particularly useful when you need to visualize the intensity of values across a 2D space, like correlation matrices, geographic maps, or any situation where the distribution or concentration of values matters. They are used in exploratory data analysis to identify patterns or correlations.

.What does the term “vectorized operation” mean in NumPy?

ANS-A vectorized operation refers to applying operations to entire arrays, element-wise, without the need for explicit loops. This is done in an optimized, compiled manner, making it much faster than using loops in Python. For example, adding two arrays together in NumPy will automatically apply the addition to each element.

.How does Matplotlib differ from Plotly?

ANS-Matplotlib is a widely used Python plotting library that provides static, 2D plots, and is great for creating a wide variety of charts in a programmatically controlled manner. Plotly, on the other hand, focuses on interactive plots. It allows for more dynamic, web-friendly visualizations (e.g., zooming, panning), making it ideal for dashboards and applications.

.What is the significance of hierarchical indexing in Pandas?

ANS-Hierarchical indexing (or MultiIndex) in Pandas allows you to have multiple levels of indexing in your DataFrame or Series. This is useful for working with high-dimensional data in a lower-dimensional form, enabling more complex data manipulation, such as grouped or nested data, without needing to reshape the dataset.

.What is the role of Seaborn’s pairplot() function?

ANS-The pairplot() function in Seaborn is used to visualize pairwise relationships between features in a dataset. It plots all numeric columns against each other in a matrix of scatterplots and optionally includes histograms or KDE plots on the diagonals. It is particularly useful for exploring correlations and interactions between multiple variables.

.What is the purpose of the describe() function in Pandas?

ANS-The describe() function provides a quick summary of the statistical properties of the numerical columns in a DataFrame. This includes metrics like the mean, standard deviation, minimum, maximum, and quartiles. It's useful for getting an overview of your data before diving into deeper analysis.

.Why is handling missing data important in Pandas?

ANS-Handling missing data is crucial because it can distort your analysis and lead to incorrect conclusions. Pandas provides various functions (e.g., .fillna(), .dropna(), and .isna()) to handle missing data, allowing you to clean and preprocess your dataset before performing analyses or building models.

.What are the benefits of using Plotly for data visualization?

ANS-Plotly offers interactive, web-ready visualizations, making it an excellent choice for dashboards, presentations, or exploratory data analysis. It supports a wide range of chart types and allows users to zoom, pan, and hover over data points to get more information. Plotly also integrates with other tools like Dash for building interactive web apps.

.How does NumPy handle multidimensional arrays?

ANS-NumPy handles multidimensional arrays (like matrices and tensors) using the ndarray object. Operations on multidimensional arrays are optimized and support element-wise operations, broadcasting, and efficient manipulation across multiple axes (dimensions).

.What is the role of Bokeh in data visualization?

ANS-Bokeh is another interactive visualization library, similar to Plotly, but with a focus on creating highly customizable and interactive plots for web applications. It supports server-based visualizations, integrates well with other Python libraries, and is ideal for generating dashboards and web apps that require real-time data interaction.

.Explain the difference between apply() and map() in Pandas?

ANS-apply(): The apply() function is used to apply a function along the axis (rows or columns) of a DataFrame or on a Series. It's more flexible and can handle a variety of operations, including row-wise or column-wise aggregations, transformations, and custom functions.

Example: df['column'].apply(lambda x: x + 1)

map(): The map() function is specifically used with Series and is more efficient for element-wise transformations. It’s mainly used to map values from one set to another (like applying a dictionary to map values) or applying simple functions to each element of the Series.

Example: df['column'].map({1: 'A', 2: 'B'})

In short, apply() is more general and works across DataFrames and Series, while map() is more specialized for Series and works best for element-wise operations.

.What are some advanced features of NumPy?

ANS- Linear Algebra: NumPy provides a full set of linear algebra functions (e.g., np.dot() for matrix multiplication, np.linalg.inv() for matrix inversion).

Random Sampling: With np.random, NumPy allows for random number generation, from simple random numbers to more advanced distributions (e.g., np.random.normal(), np.random.randint()).

Fourier Transforms: NumPy offers functions like np.fft for performing Fourier transforms, useful in signal processing.

Advanced Array Indexing: You can use techniques like boolean indexing, fancy indexing, and masking to work with arrays in advanced ways.

Broadcasting: Allows NumPy to handle operations on arrays of different shapes, enabling efficient and compact code for operations that would otherwise require loops.

.How does Pandas simplify time series analysis?

ANS-Datetime Indexing: Pandas has built-in support for handling date and time data through the datetime class, which simplifies operations like extracting year, month, day, and time components.

Resampling: The resample() method allows you to downsample or upsample time series data to different frequencies (e.g., from daily to monthly or vice versa).

Time-based indexing: You can index time series data with DatetimeIndex or PeriodIndex, making it easy to perform time-based slicing and filtering.

Rolling and Expanding Window: Pandas provides functions like rolling() and expanding() for calculating moving averages and other window-based statistics.

Time Zone Handling: Pandas makes it easy to work with time zones, converting between time zones and handling daylight saving time.

.What is the role of a pivot table in Pandas?

ANS-A pivot table in Pandas is used to summarize and aggregate data. It is similar to a pivot table in Excel, where you can rearrange and aggregate data points based on one or more keys. The pivot_table() function allows you to group data and compute statistical summaries (e.g., sum, mean) for different categories.

Example: df.pivot_table(values='sales', index='date', columns='region', aggfunc='sum')

This groups the data by date and region, summing the sales values.

.Why is NumPy’s array slicing faster than Python’s list slicing?

ANS-NumPy’s array slicing is faster because NumPy arrays are stored in a contiguous block of memory, making element access efficient. In contrast, Python lists are more flexible, which means they can store different data types, but this flexibility comes at the cost of slower access and slicing. NumPy is optimized for numerical computations and allows for better memory management and operations, especially when slicing large arrays.

.What are some common use cases for Seaborn?

ANS- Seaborn is widely used for visualizing statistical relationships in datasets. Some common use cases include:

Distribution plots: Showing distributions of data (e.g., histograms, kernel density plots). Example: sns.histplot(data)

Correlation heatmaps: Visualizing the correlation between variables in a matrix form. Example: sns.heatmap(corr_matrix)

Boxplots and violin plots: To compare distributions across categorical variables. Example: sns.boxplot(x='category', y='value')

Pair plots: To visualize pairwise relationships between multiple features. Example: sns.pairplot(data)

Categorical plots: Plotting data points in a category-based visual, such as bar plots, count plots, and strip plots. Example: sns.countplot(x='category')

Regression plots: Visualizing linear or non-linear relationships between variables. Example: sns.regplot(x='x_var', y='y_var')

```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [24, 27, 22, 32, 29],
    'Salary': [50000, 54000, 48000, 60000, 52000]
}

df = pd.DataFrame(data)


mean_age = df['Age'].mean()


print("Mean of the 'Age' column:", mean_age)


mean_salary = df['Salary'].mean()

print("Mean of the 'Salary' column:", mean_salary)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


data = {
    'Math': [90, 80, 70, 60, 75],
    'Science': [88, 76, 66, 72, 80],
    'English': [70, 60, 65, 75, 85],
    'History': [80, 70, 72, 68, 75]
}


df = pd.DataFrame(data)


corr_matrix = df.corr()

plt.figure(figsize=(8, 6))  # Optional: Set figure size
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)


plt.title('Correlation Matrix Heatmap')


plt.show()
```

```python
import numpy as np

A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

C = np.dot(A, B)  # or equivalently: C = A @ B

print("Matrix A:")
print(A)

print("\nMatrix B:")
print(B)

print("\nMatrix multiplication result (A x B):")
print(C)
```

```python
import plotly.graph_objects as go

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 11, 12, 13, 14]
z = [5, 6, 7, 8, 9]

fig = go.Figure(data=[go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='markers',
    marker=dict(
        size=8,
        color=z,
        colorscale='Viridis',
        opacity=0.8
    )
)])

fig.update_layout(
    title='3D Scatter Plot Example',
    scene=dict(
        xaxis_title='X Axis',
        yaxis_title='Y Axis',
        zaxis_title='Z Axis'
    )
)

fig.show()
```
